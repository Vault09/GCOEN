```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img =  plt.imread("images.jpeg")
```

```python
plt.imshow(img)
plt.show()
```

```python
face_detector = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces = face_detector.detectMultiScale(img, 1.1, 5)
```

```python
faces
```

```python
for f in faces:
    x,y,w,h = f
    img = cv2.rectangle(img, (x, y), (x+w, y+h), (255,255,255), 2)

plt.imshow(img)
plt.show()
```

# Web Cam

```python
%matplotlib
```

```python
cam  = cv2.VideoCapture(0)
face_detector = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")

while True:
    ret, frame = cam.read()
    if ret == False:
        continue
    
    faces = face_detector.detectMultiScale(frame, 1.2, 5)
    for f in faces:
        x,y,w,h = f
        frame = cv2.rectangle(frame, (x, y), (x+w, y+h), (255,255,255), 2)

    cv2.imshow("Face Detection", frame)
    
    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break
        
cam.release()
cv2.destroyAllWindows()
```
