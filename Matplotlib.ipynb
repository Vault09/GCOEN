```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
x_squared = x**2
```

```python
## Line Graph
```

```python
plt.plot(x,x, marker="*", label = "y=x")
plt.plot(x, x_squared, marker ="o", label= "y= x**2" )
plt.xlabel("Numbers")
plt.ylabel("Graphs")
plt.legend()
plt.title("Comparisons")
plt.show()
```

# Bar Graph

```python
x = np.array([1,2,3,4,5])

sales_in_2018 = np.array([5,3,1,8,9])
sales_in_2019 = np.array([2,6,9,1,7])
```

```python
plt.bar(x, sales_in_2018, width=0.3)
plt.bar(x+0.4, sales_in_2019, width=0.3)
plt.show()
```

# Working with Images

```python
img = plt.imread("./cat.3.jpg")
```

```python
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
cat_face =  img [  :200 ,  0: 200, :]
```

```python
plt.imshow(cat_face)
plt.show()
```

```python
cat_face = cat_face.copy()
```

```python
cat_face[ : , : , 1] = 0
```

```python
plt.imshow(cat_face)
```

```python
plt.imsave("cat_face.jpg", cat_face)
```
